```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv ("mod_df")
df.head()
```

```python
plt.figure(dpi=100)
plt.hist(df['Flat Area (in Sqft)'],bins=150)
plt.show()
```

```python
def log_transform(df):
    return np.log(df)
```

```python
df['trans_flat area']=df['Flat Area (in Sqft)'].map(log_transform)
```

```python
plt.figure(dpi=100)
plt.hist(df['trans_flat area'],bins=150)
plt.show()
```
